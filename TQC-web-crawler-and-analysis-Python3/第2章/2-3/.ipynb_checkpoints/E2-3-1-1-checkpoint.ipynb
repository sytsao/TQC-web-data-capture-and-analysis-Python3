# requests: an alternative to urllib
**requests has more functions to replace urlopen. Use request.get() to replace urlopen() and pass some parameters to the webpage.  The webbrowser is to open the new url and give you an visualization of this result.**

```python
import requests
import webbrowser
param = {"wd": "網頁資料擷取與分析特訓教材"}
r = requests.get('http://localhost:8888/view/TQC%2B%20網頁資料擷取與分析特訓教材/第2章/2-3/E2-3-1-1-input.html', params=param)
print(r.url)
webbrowser.open(r.url),encoding='utf-8'
```

## post

**We test the post function in this [page](http://pythonscraping.com/pages/files/form.html). To pass some data to the server to analyse and send some response to you accordingly.**

```python
data = {'firstname': '祥雲', 'lastname': '曹'}
r = requests.post('http://localhost:8888/view/TQC%2B%20網頁資料擷取與分析特訓教材/第2章/2-3/E2-3-1-1-input.html', data=data)
print(r.text)
```

## upload image

**We still use post function to update image in this [page](http://pythonscraping.com/files/form2.html).**

```python
file = {'uploadFile': open('./PIC_0390.jpg', 'rb')}
r = requests.post('http://localhost:8888/view/TQC%2B%20網頁資料擷取與分析特訓教材/第2章/2-3/E2-3-1-1-input.html', files=file)
print(r.text)
```

## login

**Use post method to login to a [website](http://pythonscraping.com/pages/cookies/login.html).**

```python
payload = {'username': 'test', 'password': 'password'}
r = requests.post('http://localhost:8888/view/TQC%2B%20網頁資料擷取與分析特訓教材/第2章/2-3/E2-3-1-1-input.html', data=payload)
print(r.cookies.get_dict())
r = requests.get('http://localhost:8888/view/TQC%2B%20網頁資料擷取與分析特訓教材/第2章/2-3/E2-3-1-1-input.html', cookies=r.cookies)
print(r.text)
```

## another general way to login

**Use session instead requests. Keep you in a session and keep track the cookies.**

```python
session = requests.Session()
payload = {'username': 'test', 'password': 'password'}
r = session.post('http://localhost:8888/view/TQC%2B%20網頁資料擷取與分析特訓教材/第2章/2-3/E2-3-1-1-input.html', data=payload)
print(r.cookies.get_dict())
r = session.get("http://localhost:8888/view/TQC%2B%20網頁資料擷取與分析特訓教材/第2章/2-3/E2-3-1-1-input.html")
print(r.text)
```
