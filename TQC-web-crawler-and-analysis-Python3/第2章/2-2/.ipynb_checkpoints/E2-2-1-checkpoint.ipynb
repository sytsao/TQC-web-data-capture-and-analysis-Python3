# Simplest scraping
**本範例將開啟簡單的網頁(https://www.python.org/)，接著列印其中所有的內容。**

```python
import urllib.request
with urllib.request.urlopen('http://www.python.org/') as html:
     print(html.read())
```

**若遇到網站採用utf-8 encoding編碼，則需要加上decode('utf-8')**

```python
with urllib.request.urlopen('http://www.python.org/') as html:
     print(html.read(100).decode('utf-8'))
```

若要處理可能的例外情況，例如找不到網頁或網站目前暫時關閉，需要利用urllib.error模組所回傳的資訊來處理。

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup
import sys


def getTitle(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        print(e)
        return None
    try:
        bsObj = BeautifulSoup(html, "html.parser")
        title = bsObj.body.h1
    except AttributeError as e:
        return None
    return title

title = getTitle("http://www.python.org/")
if title == None:
    print("Title could not be found")
else:
    print(title)
```

**接著利用re正則表示式對網頁標籤內容進行辨識**

```python
import re
html = urlopen("file:PYD02-1-input.html").read().decode('utf-8')
res = re.findall(r"<title>(.+?)</title>", html)
print("\nPage title is: ", res[0])
```

**另一個從網頁抓取內容選取段落的例子**

```python
res = re.findall(r"<p>(.*?)</p>", html, flags=re.DOTALL)    # re.DOTALL if multi line
print("\nPage paragraph is: ", res[0])
```

**也可以利用regex找出LINK**

```python
res = re.findall(r'href="(.*?)"', html)
print("\nAll links: ", res)
```
