```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

## 調整圖中座標軸限制

調整座標軸限制的最基本方法是使用``plt.xlim（）``和``plt.ylim（）``方法：

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

若要反向顯示任一軸，可以顛倒參數的順序：

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

``plt.axis（）``方法允許通過一次呼叫來設置``x``和``y``限制，通過一個指定``[xmin,xmax,ymin,ymax]的串列``：

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

``plt.axis（）``方法甚至允許自動收緊當前圖周圍的界限：

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

也可以確保相等的寬高比，以便在螢幕上，“x”中的一個單位等於“y”中的一個單位：

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

## 圖的標記

圖表的標籤：標題，軸標籤和簡單圖例。

標題和軸標籤是最簡單的標籤 - 有一些方法可用於快速設置它們：


```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

可以使用函數的可選參數調整這些標籤的位置，大小和樣式。

當在單個軸中顯示多條線時，創建標記每種線類型的圖例是很有用的。
同樣，Matplotlib的``plt.legend（）``方法完成，使用plot函數的``label``關鍵字指定每行的標籤：

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

``plt.legend（）``函數跟蹤線條樣式和顏色，並將它們與正確的標籤相匹配。
